pch = tipos de puntos (Plotting Character)

plotting ‘character’, i.e., symbol to use. This can either be a single character or an integer code for one of a set of graphics symbols. The full set of S symbols is available with pch = 0:18, see the examples below.

NA_integer_: no symbol.
0:18: S-compatible vector symbols.
19:25: further R vector symbols.
26:31: unused (and ignored).
32:127: ASCII characters.
128:255 native characters only in a single-byte locale and for the symbol font. (128:159 are only used on Windows.)
-32 ... Unicode code point (where supported).


```r
GenerarRpoints<-function(){
    oldPar<-par()
    par(font=2, mar=c(0.5,0,0,0))
    y=rev(c(rep(1,6),rep(2,5), rep(3,5), rep(4,5), rep(5,5)))
    x=c(rep(1:5,5),6)
    plot(x, y, pch = 0:25, cex=1.5, ylim=c(1,5.5), xlim=c(1,6.5), 
         axes=FALSE, xlab="", ylab="", bg="navy")
    text(x, y, labels=0:25, pos=3)
    par(mar=oldPar$mar,font=oldPar$font )
}

GenerarRpoints()
```

### Ejemplo : Acciones 

```r
IBM <- read.csv("IBMStock.csv")
CocaCola <- read.csv("CocaColaStock.csv")
ProcterGamble <- read.csv("ProcterGambleStock.csv")
GE <- read.csv("GEStock.csv")
Boeing <- read.csv("BoeingStock.csv")

IBM$Date = as.Date(IBM$Date, "%m/%d/%y")

GE$Date = as.Date(GE$Date, "%m/%d/%y")

CocaCola$Date = as.Date(CocaCola$Date, "%m/%d/%y")

ProcterGamble$Date = as.Date(ProcterGamble$Date, "%m/%d/%y")

Boeing$Date = as.Date(Boeing$Date, "%m/%d/%y")
```

```r
plot(CocaCola$Date, CocaCola$StockPrice, type ="l")
lines(ProcterGamble$Date, ProcterGamble$StockPrice)
```

```r
lines(ProcterGamble$Date, ProcterGamble$StockPrice)
```

```r
plot(CocaCola$Date, CocaCola$StockPrice, type ="l", col="red3")
lines(ProcterGamble$Date, ProcterGamble$StockPrice, lty=2,  col="dodgerblue3")

abline(v=as.Date(c("2000-03-01")), lwd=1)

abline(v=as.Date(c("1983-01-01")), lwd=1, col="blue4")
```

```r
plot(CocaCola$Date[301:432], CocaCola$StockPrice[301:432], type="l", col="red3", ylim=c(0,210), lwd=2, bty="n")

lines(ProcterGamble$Date[301:432], ProcterGamble$StockPrice[301:432], lty=2,lwd=2,  col="dodgerblue3")

lines(GE$Date[301:432], GE$StockPrice[301:432], lty=3,lwd=2,  col="darkblue")

lines(IBM$Date[301:432], IBM$StockPrice[301:432], lty=4, lwd=2, col="magenta4")

lines(Boeing$Date[301:432], Boeing$StockPrice[301:432], lty=5, lwd=2, col="orange")


legend("topright", legend=c("CocaCola","P&G","GE","IBM","Boeing"), 
       col=c("red3","dodgerblue3","darkblue","magenta4","orange"),
       lty = 1:5, cex=0.55, lwd=2, title="Stocks", text.font=24, bg='snow1')

abline(v=as.Date(c("2000-03-01")), lty=2)

abline(v=as.Date(c("1997-09-01")), lty=2, col="navy")
abline(v=as.Date(c("1997-11-01")), lty=2, col="navy")

abline(v=as.Date(c("2004-01-01")), lty=2, col="orange")
```

### Ejemplo: Ggplot2 - Regresión Automoviles

```r
options(repr.plot.width=7, repr.plot.height=5, encoding='UTF-8')
```

```r
head(mtcars)
```

```r
plot(mtcars$wt, mtcars$mpg)
abline(lm(mpg~wt, data=mtcars))
title("Regresión de MPG(Millas por Gálon) sobre Weight(Peso)")
```

```r
library(ggplot2)
ggplot(data=mtcars, aes(x=wt, y=mpg)) +
geom_point() +
labs(title="Datos de Automoviles", x="Peso", y="Millas por galón")
```

```r
ggplot(data=mtcars, aes(x=wt, y=mpg)) +
geom_point(pch=17, color="blue", size=2) +
geom_smooth(method="lm", color="red", linetype=2) +
labs(title="Datos de Automoviles", x="Peso", y="Millas por galón")
```

```r
mtcars$am <- factor(mtcars$am, levels=c(0,1),labels=c("Automatic", "Manual"))
mtcars$vs <- factor(mtcars$vs, levels=c(0,1),labels=c("V-Engine", "Straight Engine"))
mtcars$cyl <- factor(mtcars$cyl)
```

```r
ggplot(data=mtcars, aes(x=hp, y=mpg, 
    shape=cyl, color=cyl)) +
    geom_point(size=3) +
    facet_grid(am ~ vs) +
    labs(title="Datos de Automoviles por Tipo de Motor",
          x="Caballos de Potencia", y="Millas por Galon")
```

```r
ggplot(data=mtcars, aes(x=wt, y=mpg, shape=cyl, color=cyl)) + 
    geom_point(size=3) +
    stat_smooth(method="lm", colour="red", linetype=2) +
    labs(title="Datos de Automoviles", x="Peso", y="Millas por galón")

```
